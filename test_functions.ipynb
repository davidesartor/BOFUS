```python
from src.test_functions import *
import matplotlib.pyplot as plt
import jax
import jax.numpy as jnp
import jax.random as jr

jax.config.update("jax_enable_x64", True)

one_dim_functions = [
    # Many Local Minima
    GramacyLee(),
    # Others
    Forrester(),
]
two_dim_functions = [
    # Many Local Minima
    Bukin6(),
    CrossInTray(),
    DropWave(),
    EggHolder(),
    HolderTable(),
    Langermann(),
    Levy13(),
    Schaffer2(),
    Schaffer4(),
    Shubert(),
    # Bowl-Shaped
    Bohachevsky1(),
    Bohachevsky2(),
    Bohachevsky3(),
    # Plate-Shaped
    Booth(),
    Matyas(),
    McCormick(),
    # Valley-Shaped
    Camel3(),
    Camel6(),
    # Steep Ridges/Drops
    DeJong5(),
    Easom(),
    # Others
    Beale(),
    Branin(),
    GoldsteinPrice(),
]
n_dim_functions = [
    # Many Local Minima
    Ackley(),
    Griewank(),
    Levy(),
    Rastrigin(),
    Schwefel(),
    # Bowl-Shaped
    Perm0(),
    RotatedHyperEllipsoid(),
    Sphere(),
    SumPowers(),
    SumSquares(),
    Trid(),
    # Plate-Shaped
    PowerSum(),
    Zakharov(),
    # Valley-Shaped
    DixonPrice(),
    Rosenbrock(),
    # Steep Ridges/Drops
    Michalewicz(),
    #Others
    Perm(),
    StyblinskiTang(),
]
specific_dim_functions = [
    (4, Colville()),
    (3, Hartmann3()),
    (4, Hartmann4()),
    (6, Hartmann6()),
    (4, Powell()),
    (4, Shekel()),
]
print(
    f"total: {len(one_dim_functions) + len(two_dim_functions) + len(n_dim_functions)+ len(specific_dim_functions)} functions"
)
```

# 1-dimensional functions

```python
for test_fn in one_dim_functions:
    X = jnp.linspace(0, 1, 100000)[:, None]
    Y = test_fn(X)

    plt.figure(figsize=(6, 4))
    plt.plot(X[:, 0], Y)
    plt.hlines(Y.min(), X.min(), X.max(), colors="red", label=f"min: {Y.min():.4f}")
    plt.legend()
    plt.title(f"{test_fn.__class__.__name__}")
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.grid()
    plt.show()
```

# 2-dimensional functions

```python
for test_fn in two_dim_functions:
    x1 = jnp.linspace(0, 1, 1000)
    x2 = jnp.linspace(0, 1, 1000)
    X1, X2 = jnp.meshgrid(x1, x2)
    X = jnp.stack([X1, X2], axis=-1)
    Y = test_fn(X)

    plt.figure(figsize=(4, 4))
    plt.contourf(X1, X2, Y, levels=50, cmap="viridis")
    plt.plot(*X[jnp.unravel_index(jnp.argmin(Y), X.shape[:-1])], "rx", label=f"min: {Y.min():.4f}")
    plt.legend()
    plt.title(f"{test_fn.__class__.__name__}")
    plt.colorbar()
    plt.xlabel("x1")
    plt.ylabel("x2")
    plt.gca().set_aspect("equal", adjustable="box")
    plt.show()
```

# n-dimensional functions

```python
for test_fn in n_dim_functions:
    x1 = jnp.linspace(0, 1, 1000)
    x2 = jnp.linspace(0, 1, 1000)
    X1, X2 = jnp.meshgrid(x1, x2)
    X = jnp.stack([X1, X2], axis=-1)
    Y = test_fn(X)

    plt.figure(figsize=(4, 4))
    plt.contourf(X1, X2, Y, levels=50, cmap="viridis")
    plt.plot(*X[jnp.unravel_index(jnp.argmin(Y), X.shape[:-1])], "rx", label=f"min: {Y.min():.4f}")
    plt.legend()
    plt.title(f"{test_fn.__class__.__name__}")
    plt.colorbar()
    plt.xlabel("x1")
    plt.ylabel("x2")
    plt.gca().set_aspect("equal", adjustable="box")
    plt.show()
```
